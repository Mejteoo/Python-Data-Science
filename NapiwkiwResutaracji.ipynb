```python
import numpy as np
import seaborn as sns

df = sns.load_dataset("tips")
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
sredniaday = df.groupby(['day', 'sex']).agg({'total_bill': 'mean', 'tip': 'mean'})
print(sredniaday)
```

```python
df['tip_rate'] = df['tip'] / df['total_bill']
print(df['tip_rate'])
```

```python
sns.scatterplot(data=df, x=df['total_bill'], y=df['tip_rate'], hue=df['sex'])
```
